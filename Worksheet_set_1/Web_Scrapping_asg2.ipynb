Q9: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1.First get the webpage https://www.azquotes.com/
2. Click on Top Quotes
3. Than scrap a) Quote b) Author c) Type Of Quote

```python
driver = webdriver.Chrome(r'C:\Users\Praveen\chromdriver\chromedriver.exe')
url = 'https://www.azquotes.com/'
driver.get(url)
```

```python
# clicking using absolute xpath function
search_btn = driver.find_element(By.XPATH,'//a[@href="/top_quotes.html"]')
search_btn.click()
```

```python
Top_Quote = pd.DataFrame()
Quote_ = []
Author_ = []
Type_of_Quote_ = []
start =0
end = 9 

for page in range(start,end):
    quote = driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote:
        Quote_.append(i.text)
        
    author = driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in author:
        Author_.append(i.text)
    
    Type = driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in Type:
        Type_of_Quote_.append(i.text)
        
    next_button = driver.find_element(By.XPATH,'//li[@class="next"]')
    next_button.click()
    

    a = driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in a:
        Quote_.append(i.text)
        
    b = driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in b:
        Author_.append(i.text)
    
    c = driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in c:
        Type_of_Quote_.append(i.text)

Top_Quote['Quote'] = Quote_[0:1000]
Top_Quote['Author'] = Author_[0:1000]
Top_Quote['Type_of_Quote'] = Type_of_Quote_[0:1000]
Top_Quote
```
