```python
#lets see how data is distributed for every column
plt.figure(figsize = (50,40), facecolor ='skyblue')
plotnumber = 1

for column in data:
  if plotnumber<=68:
    ax = plt.subplot(17,4,plotnumber)
    sns.boxplot(data= data[column])
    plt.xlabel(column,fontsize = 15)
  plotnumber+=1
plt.tight_layout()
```

* We have most of the  column with outliers and boxplot is clearly showing that many of them column have skewness 
* inplace of removing every columns outliers either we can normalize it , if we dont want to remove the data as the data value which has been shown in the data set are all real and possible value

# let's see the data after standardization 

```python
from sklearn.preprocessing import StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
std = StandardScaler()
scaled_data = std.fit_transform(data)
```

```python
vif = pd.DataFrame()
vif['vif'] = [variance_inflation_factor(scaled_data,i) for i in range(scaled_data.shape[1])]
vif
```
